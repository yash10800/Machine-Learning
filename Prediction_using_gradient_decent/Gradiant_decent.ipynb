```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt


data=pd.read_csv('Train.csv')
x=data.iloc[:,0]
y=data.iloc[:,1]



#y=theta0+theta1(x)
#m=theta1 c=theta0
m,c=0,0
alfa=0.001
niter=1000
n=len(x)
for i in range(niter):
    y1=m*x + c
    dm=(-2/float(n)) * sum( x * (y-y1))
    dc=(-2/float(n)) * sum(y-y1)
    m= m - (alfa*dm)
    c= c - (alfa*dc)
print(m,c)
y1=m*x+c
plt.scatter(x,y)
plt.scatter(x,y1)
plt.show()
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt


data = pd.read_csv('Train.csv')
X = data.iloc[:, 0]
Y = data.iloc[:, 1]
plt.scatter(X, Y)
plt.show()

m = 0
c = 0

L = 0.0001  
epochs = 1000 

n = float(len(X)) 


for i in range(epochs): 
    Y_pred = m*X + c  
    D_m = (-2/n) * sum(X * (Y - Y_pred))  
    D_c = (-2/n) * sum(Y - Y_pred)  
    m = m - L * D_m  
    c = c - L * D_c  
    
print (m, c)

Y_pred = m*X + c

plt.scatter(X, Y)
plt.plot([min(X), max(X)], [min(Y_pred), max(Y_pred)], color='red')
plt.show()
```
